#1. 다음을 출력하는 프로그램을 작성하시오.

> **100** little **monkeys** jumping on the bed.

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, "No more monkeys jumping on the bed!"

> ...

> ... 

> **1** little **monkey** jumping on the bed. 

> One fell off and broke his head. 

> **Mama(or Daddy)** called the doctor and the doctor said, 

> "No more monkeys jumping on the bed!" 

(참고) Mama 혹은 Daddy를 랜돔하게 선택하게 하자.

```python
# 여기에 코드를 입력하세요.
import random
parents = ["Mama", "Daddy"]
gender = ["his", "her"]

number = 100
while number > 1:
    print("{} little monkeys jumping on the bed.".format(number))
    print("One fell off and broke his head.")
    parent = random.choice(parents)
    print('{} called the doctor and the doctor said,'.format(parent))
    print('"No more monkeys jumping on the bed!"')
    number -= 1
    if number == 1:
        print("{} little monkey jumping on the bed.".format(number))
        gender_select = random.choice(gender)
        print("One fell off and broke {} head.".format(gender_select))
        parent = random.choice(parents)
        print('{} called the doctor and the doctor said,'.format(parent))
        print('"No more monkeys jumping on the bed!"')
```

#2. 시작하는 연도와 끝나는 연도를 물어보는 프로그램을 작성 하고, 그 사이에 있는 모든 윤년을 출력하라. 

(시작하는 연도와 끝나는 연도가 윤년이면 그 해도 출력) 

* 윤년은 4로 나누어짐 (예: 1984, 2012)
* 100으로 나누어지면 윤년이 아님 (예: 1800, 1900) 
* 윤년은 400으로 나누어짐 (예: 1600, 2000) 

> Pick a starting year (like 1973 or something): 1973

> Now pick an ending year: 1977 

> Check it out... these years are leap years: 1976 

(참고) 윤년 리스트: http://mwultong.blogspot.com/2005/12/18002300-leap-year-list.html

```python
# 여기에 코드를 입력하세요.
start_year = input("Pick a starting year:")
finish_year = input("Now pick an ending year:")

start_year = int(start_year)
finish_year = int(finish_year)
leapyears = []

for year in range(start_year-1,finish_year+1):
    if year % 4 != 0:
        year += 1
    elif year % 100 == 0:
        if year % 400 != 0:
            year += 1
        else:
            leapyears.append(year)
            year +=1
    else:
        leapyears.append(year)
        year += 1


print("Check it out... these years are leap years: "+ ", ".join([str(i) for i in leapyears] ))
```

#3. 구구단을 출력하는 프로그램을 작성하고 결과를 txt 파일로 저장하시오.

```python
# 여기에 코드를 입력하세요.
gugudan = []

for i in range(1,10):
    tmp = []
    for j in range(1,10):
        multiply = str(i * j)
        tmp_str = "{} x {} = ".format(i, j) + multiply
        tmp.append(tmp_str)
        j += 1
    gugudan.append(tmp)
    i += 1

for i in range(9):
        gugudan[i]="  ".join(element for element in gugudan[i])
        gugudan[i] = str(i+1)+"단 : " +gugudan[i]

print("""
""".join([str(i) for i in gugudan]))


f = open("result.txt", "w") 
for i in gugudan:
    f.write(str(i) + "\n")
f.close()
```
