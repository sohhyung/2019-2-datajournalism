# 텍스트 전처리

### 전처리 함수들 모음

```python
# 앞 부분기자 이름 없애기 + 따옴표 변환 함수
def del_writer(text):
    text = text.replace('“', '"')
    text = text.replace('”', '"')
    text = text.replace("‘", "'")
    text = text.replace("’", "'")
    if text.strip()[0] == '[':
        try:
            text = text.split('기자]')[1].strip()
        except:
            try:
                text = text.split('에디터]')[1].strip()
            except:
                text = text
    return text
```

```python
# 뉴센 뒷 부분 : 뉴스엔 김명미 mms2@기사제보 및 보도자료 newsen@newsen.comcopyrightⓒ 뉴스엔. 무단전재 & 재배포 금지
def del_ns(text):
    result = ''
    split_text = text.split('뉴스엔')
    # 보통의 경우
    if len(split_text) == 3:
        result = split_text[0]
    # 기사 내용에 '뉴스엔' 포함될 시
    else:
        for st in split_text[:-2]:
            result += st + "뉴스엔"
        # 마지막 '뉴스엔 제거'
        result = result[:-3]
    return result
```

```python
# 헤럴드팝 뒷 부분 : popnews@heraldcorp.com▶ K-POP 한류스타 바로 앞에서 본다 \'리얼직캠\'[헤럴드팝 네이버TV]- Copyrights ⓒ 헤럴드POP & heraldpop.com, 무단 전재 및 재배포 금지 -'
def del_hp(text):
    text = text.split('popnews')[0].strip()
    return text
```

```python
# 스타뉴스 뒷 부분
# 1안 공미나 기자 mnxoxo@mtstarnews.com▶ 스타뉴스 단독 ▶ 생생 스타 현장▶ 스타 인기영상 <저작권자 ⓒ ‘리얼타임 연예스포츠 속보,스타의 모든 것’ 스타뉴스, 무단전재 및 재배포 금지>
# 2안 김미화 기자 letmein@▶ 스타뉴스 단독 ▶ 생생 스타 현장▶ 스타 인기영상 <저작권자 ⓒ ‘리얼타임 연예스포츠 속보,스타의 모든 것’ 스타뉴스, 무단전재 및 재배포 금지>
# 3안 ▶ 스타뉴스 단독 ▶ 생생 스타 현장▶ 스타 인기영상 <저작권자 ⓒ ‘리얼타임 연예스포츠 속보,스타의 모든 것’ 스타뉴스, 무단전재 및 재배포 금지>

# 우선 세모 뒤만 삭제
def del_st(text):
    text = text.split('▶')[0].strip()
    return text
```

## 포토 & 방송 기사 레이블링

```python
# 헤럴드팝 [포토] |[어게인TV] [투데이TV]
# 뉴센 [포토엔HD] |[오늘TV]
# 오센 [사진]
# 스타뉴스 [★포토] |[별별TV][★밤TV]
# SBS funE [E포토]|[스브스夜]
```

```python
# 포토 찾아 return
photos = ["포토]", "[포토", "[사진"]
def is_photo(title):
    if any(p in title for p in photos) :
        result = 1
    else:
        result = 0
    return result
```

```python
# 함수 확인
title = ns_df.iloc[4995,0]
print(title)
print(is_photo(title))
```

```python
# 방송 찾아 return
tvs = ["TV]", "[스브스夜]"]
def is_tv(title):
    if any(t in title for t in tvs) :
        result = 1
    else:
        result = 0
    return result
```

```python
# 함수 확인
title = ns_df.iloc[4995,0]
print(title)
print(is_tv(title))
```

```python
# 모든 dataframe photo, tv 행 추가
for df in df_list:
    df['photo'] = df['title'].apply(lambda x : is_photo(x))
for df in df_list:
    df['tv'] = df['title'].apply(lambda x : is_tv(x))
```

## SNS 기사 레이블링

```python
# SNS 단어 찾아 return
def find_sns(text):
    sns = " "
    if '인스타그램' in text:
        sns += '인스타그램 '
    if '트위터' in text:
        sns += '트위터 '
    if '유튜브' in text:
        sns += '유튜브 '
    sns = sns.strip()
    return sns
```

```python
# 함수 확인
text = hp_df.iloc[10,1]
print(text)
print(find_sns(text))
```

```python
# 모든 dataframe SNS 행 추가
for df in df_list:
    df['sns'] = df['text'].apply(lambda x: find_sns(x))
```

```python
from collections import Counter

for num, df in enumerate(df_list):
    print(df_name[num])
    c = Counter(df['sns'])
    print(c)
```

## 익명 기사 레이블링

```python
# 익명 단어 찾아 return
def find_ano(text):
    anonymity = ["네티즌", "A씨는", "익명", "관계자", "누리꾼", "측근", "최측근"]
    anos = ""
    for ano in anonymity:
        if ano in text:
            anos += ano + " "
    anos = anos.strip()
    return anos
```

```python
# 함수 확인
text = hp_df.iloc[766,1]
print(text)
print(find_ano(text))
```

```python
# anonymous 행 추가
for df in df_list:
    df['ano'] = df['text'].apply(lambda x: find_ano(x))
```

```python
from collections import Counter

for num, df in enumerate(df_list):
    print(f"< {df_name[num]} >")
    c = Counter(df['ano'])
    print(c)
```
